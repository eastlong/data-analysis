```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# np.linspace : 在指定的间隔内返回均匀间隔的数字。
# 返回num均匀分布的样本，在[start, stop]。
plot_x = np.linspace(-1., 6., 141)
plot_x
```

```python
plot_y= (plot_x - 2.5)**2 - 1 
```

```python
plt.plot(plot_x, plot_y)
plt.show()
```

```python
# 表示一个很小的数
epsilon = 1e-8
# 学习率
eta = 0.1
```

```python
# J(theta) 表示y
def J(theta):
    return (theta-2.5)**2 - 1.

# 求导数
def dJ(theta):
    return 2*(theta - 2.5)

theta = 0.0

while True:
    gradient = dJ(theta)
    last_theat = theta 
    theta = theta - eta * gradient

    if (abs(J(theta) - J(last_theat)) < epsilon):
        break

print(theta)
print(J(theta)) # J(theat)为最小值，即截距-1
```

```python
theta = 0.0
theta_history = [theta]
while True:
    gradient = dJ(theta)
    last_theat = theta 
    theta = theta - eta * gradient
    theta_history.append(theta)

    if (abs(J(theta) - J(last_theat)) < epsilon):
        break

plt.plot(plot_x, J(plot_x))
plt.plot(np.array(theta_history), J(np.array(theta_history)), color='r',marker='+')
plt.show()
```

```python
def J(theta):
    try:
        return (theta-2.5)**2 - 1.
    except:
        return float('inf')
```

```python
# 设置最大遍历次数

def gradient_descent(initial_theta, eta, n_iters = 1e4, epsilon=1e-8):
    
    theta = initial_theta
    i_iter = 0
    theta_history.append(initial_theta)

    while i_iter < n_iters:
        gradient = dJ(theta)
        last_theta = theta
        theta = theta - eta * gradient
        theta_history.append(theta)
    
        if(abs(J(theta) - J(last_theta)) < epsilon):
            break
            
        i_iter += 1
        
    return
```

```python
eta = 1.1
theta_history = []
gradient_descent(0, eta)
print(len(theta_history))
```
